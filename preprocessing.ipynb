# Import module

```python
import os
import re
import pandas as pd
import tensorflow as tf
from underthesea import word_tokenize
import numpy as np
```

```python
from transformers import AutoTokenizer
# Tokenizer để encode đầu vào văn bản
model_pretrain = 'vinai/phobert-base'
tokenizer = AutoTokenizer.from_pretrained(model_pretrain)
```

# Tạo các hàm tiền xử lý

```python
def remove_special_char(text):
    res = re.sub(r'[!#\()*+;<=>[\]^`{|}~]', "", text)
    res = re.sub(r'\s+', ' ', res).strip()
    return res

def process_en(text):
    res = text.lower()
    res = re.sub(r"i'm", "i am", res)
    res = re.sub(r"he's", "he is", res)
    res = re.sub(r"she's", "she is", res)
    res = re.sub(r"it's", "it is", res)
    res = re.sub(r"that's", "that is", res)
    res = re.sub(r"what's", "what is", res)
    res = re.sub(r"where's", "where is", res)
    res = re.sub(r"how's", "how is", res)
    res = re.sub(r"\'ll", " will", res)
    res = re.sub(r"\'ve", " have", res)
    res = re.sub(r"\'re", " are", res)
    res = re.sub(r"\'d", " would", res)
    res = re.sub(r"\'re", " are", res)
    res = re.sub(r"won't", "will not", res)
    res = re.sub(r"can't", "cannot", res)
    res = re.sub(r"n't", " not", res)
    res = re.sub(r"n'", "ng", res)
    res = re.sub(r"'bout", "about", res)
    res = re.sub(r"'til", "until", res)
    res = remove_special_char(res)
    return res

def process_vi(text):
    res = text.lower()
    res = remove_special_char(res)
    res = word_tokenize(res, format='text')
    return res
```

# TIền xử lý

```python
# đọc dữ liệu từ file txt
ROOT_PATH = 'dataset/'
en = []
with open(ROOT_PATH+'title_en.txt', 'r', encoding='utf8') as f:
    for line in f.readlines():
        en.append(process_en(line))
vi = []
with open(ROOT_PATH+'title_vi.txt', 'r', encoding='utf8') as f:
    for line in f.readlines():
        vi.append(process_vi(line))
```

```python
df = pd.DataFrame([data for data in zip(en, vi)], columns=['en', 'vi'])
df
```

```python
df.to_csv('translate.csv')
```
